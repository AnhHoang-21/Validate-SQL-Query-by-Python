## Extract all VIP customers

```python
vip_customers = customers[customers['VIP_customer'] == 'yes']

# Display the resulting dataframe
vip_customers.head()
```

## Extract first three months data in the sales table

```python
# Convert 'SaleDate' to datetime
sales['SaleDate'] = pd.to_datetime(sales['SaleDate'])

# Extract rows for the first three months
first_three_months = sales[(sales['SaleDate'] >= '2024-07-01') & (sales['SaleDate'] < '2024-10-01')]

# Display the resulting dataframe
first_three_months.head()
```

## Total sales calculation with a loop
**Totals sales of all VIP customers in the first three months**

```python
total_price = 0

for i in vip_customers["CustomerID"]:
    specific_customer_rows = first_three_months[first_three_months['CustomerID'] == i]
    total_price += specific_customer_rows['TotalPrice'].sum()

print("Total Price for VIP customers:", total_price)
```
